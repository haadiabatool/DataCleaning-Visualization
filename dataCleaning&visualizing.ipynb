Missing Values Filled With Mode(Most Frequently Occuring Values)

```python
data_set['Teacher_Quality'].fillna(data_set['Teacher_Quality'].mode()[0],inplace=True)
```

```python
data_set['Parental_Education_Level'].fillna(data_set['Parental_Education_Level'].mode()[0],inplace=True)
```

```python
data_set['Distance_from_Home'].fillna(data_set['Distance_from_Home'].mode()[0],inplace=True)
```

Data Visualization

```python
# Hours Studied vs Exam Score
sns.scatterplot(x='Hours_Studied',y='Exam_Score',data=data_set)

plt.show()
```

```python
# Exam Score Distribution
plt.figure(figsize=(8,6))
sns.histplot(data_set['Exam_Score'],bins=25,color='Brown')
plt.title('Distribution of Exam Score')
plt.show()
```

```python
# Access to Resources vs Exam Score Distribution
sns.barplot(x='Access_to_Resources', y='Exam_Score', data=data_set)

plt.title("Average Exam Score by Access to Resources")
plt.show()
```

```python
# Gender Distribution
fig, ax=plt.subplots(1,2,figsize=(8,5))
gender_counts=data_set['Gender'].value_counts()
ax[0].pie(gender_counts,labels=gender_counts.index,autopct='%1.1f%%')
ax[0].set_title('Pie Chart')

sns.countplot(x='Gender',data=data_set,ax=ax[1])
fig.suptitle('Gender Distribution')
ax[1].set_title('Count Plot')
plt.tight_layout()
plt.show()
```

```python
# Motivational Level Distribution
sns.countplot(x='Motivation_Level',data=data_set)
plt.title('Motivation level Distribution')
plt.show()
```

```python
# Numeric Columns Distributions 
columns=['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores','Tutoring_Sessions','Physical_Activity','Exam_Score']
axes=data_set[columns].hist(figsize=(20,18))
for ax,col in zip(axes.flatten(),columns):
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')
    ax.set_title(col+" Distribution")


plt.suptitle("Numeric Columns Distribution")
plt.tight_layout()
plt.show()
```

```python
# Numeric Features Vs Exam Score
columns=['Hours_Studied','Attendance','Sleep_Hours','Previous_Scores','Tutoring_Sessions','Physical_Activity','Exam_Score']

fig,axes=plt.subplots(3,3,figsize=(18,10))
axes=axes.flatten()

for i,col in enumerate(columns):
    sns.scatterplot(x=data_set[col],y=data_set['Exam_Score'], ax=axes[i])

    axes[i].set_title(col+' vs Exam Score')
    axes[i].set_xlabel(col)
    axes[i].set_ylabel('Exam Score')

for j in range(len(columns),len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Numeric Feature vs Exam Score')
plt.tight_layout()
plt.show()
```

```python
# Categorical Features vs Exam Scores
categorical_columns=['Parental_Involvement','Access_to_Resources','Extracurricular_Activities','Motivation_Level',
                     'Internet_Access','Family_Income','Teacher_Quality','School_Type',
                     'Peer_Influence','Learning_Disabilities','Parental_Education_Level','Distance_from_Home','Gender']
fig,axes=plt.subplots(5,3,figsize=(20,16))
axes=axes.flatten()
for i,col in enumerate(categorical_columns):
    sns.boxplot(x=col,y='Exam_Score',data=data_set,ax=axes[i])
    axes[i].set_title(col+ " Vs Box Plot")
    axes[i].set_ylabel("Exam Score")

for j in range(len(categorical_columns),len(axes)):
    axes[j].set_visible(False)
plt.suptitle('Numeric Columns Boxplots')
plt.tight_layout()
plt.show()
```

```python
# Correlation Heatmap
corr=data_set.corr(numeric_only=True)
plt.figure(figsize=(8,6))
sns.heatmap(corr,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
